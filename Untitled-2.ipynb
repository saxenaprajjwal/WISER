```python
%matplotlib inline
# 1-Layer Quantum Galton Board (Quantum Peg)
import numpy as np
import matplotlib.pyplot as plt
from qiskit import QuantumCircuit, ClassicalRegister, QuantumRegister
# Updated import for V2 Primitives
from qiskit.primitives import StatevectorSampler
from qiskit.visualization import plot_histogram

# --- 1. Define Circuit Parameters ---

# For a 1-layer Galton board, we have 1 level of pegs.
# This results in 2 possible output bins.
layers = 1
num_output_bins = layers + 1

# According to the paper's model (Fig. 3), a single peg requires:
# - 1 control qubit
# - 3 working qubits (1 for the initial 'ball', 2 for the output channels)
# Total qubits = 4
q_control = QuantumRegister(1, 'control')
q_board = QuantumRegister(3, 'board')

# We need classical registers to store the measurement results of the output bins.
c_output = ClassicalRegister(num_output_bins, 'c')

# Create the quantum circuit
qc = QuantumCircuit(q_control, q_board, c_output)

# --- 2. Build the Quantum Peg Circuit ---
# This sequence of gates mimics a ball hitting a peg and having a 50/50
# chance of going left or right.

# The 'board' qubits are indexed q_board[0], q_board[1], q_board[2].
# Let's map them to the paper's q1, q2, q3 for clarity.
# q_board[0] -> q1 (left output channel)
# q_board[1] -> q2 (input 'ball' channel)
# q_board[2] -> q3 (right output channel)

# Place the 'ball' on the middle qubit (q_board[1])
qc.x(q_board[1])
qc.barrier()

# Put the control qubit into superposition. This is what creates the
# probabilistic nature of the 'peg'.
qc.h(q_control[0])
qc.barrier()

# First controlled-SWAP (Fredkin gate). If control is |1>, the ball
# moves from the center to the left channel.
qc.cswap(q_control[0], q_board[0], q_board[1])

# An 'inverted' CNOT to prepare the control qubit for the next step.
qc.cx(q_board[1], q_control[0])

# Second controlled-SWAP. If control is |1>, the ball
# moves from the center to the right channel.
qc.cswap(q_control[0], q_board[1], q_board[2])
qc.barrier()

# --- 3. Measure the Output Bins ---
# We measure the two output channel qubits to see where the 'ball' landed.
# The result '01' means the ball is in the left bin (board[0]).
# The result '10' means the ball is in the right bin (board[2]).
qc.measure(q_board[0], c_output[0])
qc.measure(q_board[2], c_output[1])

# --- 4. Display and Simulate the Circuit ---

print("--- Quantum Circuit for 1-Layer Galton Board ---")
# Using the 'mpl' style for a graphical circuit diagram.
display(qc.draw('mpl'))

print("\n--- Simulating the Circuit (using V2 Primitives) ---")
# Using the StatevectorSampler (V2) to run the circuit.
sampler = StatevectorSampler()
# The V2 run method now requires a list of circuits, so we wrap qc in [].
job = sampler.run([qc], shots=8192)
result = job.result()
# In V2, counts are accessed from the data attribute of the result.
counts = result[0].data.c.get_counts()


print("\n--- Simulation Results ---")
print(counts)

# --- 5. Plot the Results ---
# The plot should show a roughly 50/50 distribution between the two
# possible outcomes: '01' and '10'.
plot_histogram(counts, title='1-Layer Quantum Galton Board Simulation Results')
plt.xlabel("Output Bin (Right-to-Left)")
plt.ylabel("Frequency (Counts)")
plt.show()
```

```python

# 2-Layer Quantum Galton Board
import numpy as np
import matplotlib.pyplot as plt
from qiskit import QuantumCircuit, ClassicalRegister, QuantumRegister
from qiskit.primitives import StatevectorSampler
from qiskit.visualization import plot_histogram

# --- 1. Define Circuit Parameters ---

# For a 2-layer Galton board, we have 2 levels of pegs.
# This results in 3 possible output bins.
layers = 2
num_output_bins = layers + 1

# According to the paper's model, a 2-layer board requires:
# - 1 control qubit (reused)
# - 5 board qubits (1 for the initial ball, 4 for channels)
# Total qubits = 6
q_control = QuantumRegister(1, 'control')
q_board = QuantumRegister(2 * layers + 1, 'board')

# We need classical registers to store the measurement results of the output bins.
c_output = ClassicalRegister(num_output_bins, 'c')

# Create the quantum circuit
qc = QuantumCircuit(q_control, q_board, c_output)


# --- Helper Function for a single "Quantum Peg" (Hadamard removed) ---
def quantum_peg_logic(qc, control, ball_in, left_out, right_out):
    """
    Applies the core logic of a quantum peg (CSWAPs and CNOT).
    The Hadamard gate is now applied at the layer level.
    """
    # Controlled-SWAP for left path
    qc.cswap(control, left_out, ball_in)

    # Inverted CNOT
    qc.cx(ball_in, control)

    # Controlled-SWAP for right path
    qc.cswap(control, ball_in, right_out)
    qc.barrier()


# --- 2. Build the Quantum Galton Board Circuit ---

# Layer 0: Initial Ball
# Place the 'ball' on the middle qubit for the first layer.
# For 2 layers, the board qubits are 0, 1, 2, 3, 4. The middle is 2.
initial_ball_qubit_index = layers
qc.x(q_board[initial_ball_qubit_index])
qc.barrier()

# Layer 1: First row of pegs (1 peg)
print("Building Layer 1...")
# Apply H-gate ONCE for the layer
qc.h(q_control[0])
# The ball at board[2] can go to board[1] (left) or board[3] (right)
quantum_peg_logic(qc, q_control[0], q_board[2], q_board[1], q_board[3])

# *** CRUCIAL STEP: Reset the control qubit for the next layer ***
qc.reset(q_control[0])
qc.barrier()

# Layer 2: Second row of pegs (2 pegs)
print("Building Layer 2...")
# Apply H-gate ONCE for the layer
qc.h(q_control[0])
# Peg 1: Input from board[1], outputs to board[0] and board[2]
quantum_peg_logic(qc, q_control[0], q_board[1], q_board[0], q_board[2])

# This CNOT is the corrective gate shown in Fig. 4 of the paper.
# It rebalances the control qubit's probability for the next peg in the same row.
qc.cx(q_board[2], q_control[0])

# Peg 2: Input from board[3], outputs to board[2] and board[4]
quantum_peg_logic(qc, q_control[0], q_board[3], q_board[2], q_board[4])
qc.barrier()


# --- 3. Measure the Output Bins ---
# The final output bins are qubits 0, 2, and 4.
# We expect results like '001', '010', '100'.
qc.measure(q_board[0], c_output[0])
qc.measure(q_board[2], c_output[1])
qc.measure(q_board[4], c_output[2])


# --- 4. Display and Simulate the Circuit ---

print("\n--- Quantum Circuit for 2-Layer Galton Board ---")
# Using the 'clifford' style for a graphical circuit diagram.
display(qc.draw('mpl', style='clifford'))

print("\n--- Simulating the Circuit (using V2 Primitives) ---")
sampler = StatevectorSampler()
job = sampler.run([qc], shots=8192)
result = job.result()
counts = result[0].data.c.get_counts()

print("\n--- Simulation Results ---")
# The results are in binary strings. We can map them for clarity.
# '001' -> Bin 0, '010' -> Bin 1, '100' -> Bin 2
remapped_counts = {
    "Bin 0 (Left)": counts.get('001', 0),
    "Bin 1 (Center)": counts.get('010', 0),
    "Bin 2 (Right)": counts.get('100', 0)
}
print(remapped_counts)


# --- 5. Plot the Results ---
# The plot should show a 1:2:1 distribution.
plot_histogram(remapped_counts, title='2-Layer Quantum Galton Board Simulation Results')
plt.xlabel("Output Bin")
plt.ylabel("Frequency (Counts)")
plt.show()

```

```python


# --- Helper Function for a single "Quantum Peg" (Hadamard removed) ---
def quantum_peg_logic(qc, control, ball_in, left_out, right_out):
    """
    Applies the core logic of a quantum peg (CSWAPs and CNOT).
    The Hadamard gate is applied at the layer level.
    """
    # Controlled-SWAP for left path
    qc.cswap(control, left_out, ball_in)

    # Inverted CNOT
    qc.cx(ball_in, control)

    # Controlled-SWAP for right path
    qc.cswap(control, ball_in, right_out)

# --- Main Function to Create the Galton Board Circuit ---
def create_qgb_circuit(layers):
    """
    Generates a Quantum Galton Board circuit for a given number of layers.

    Args:
        layers (int): The number of layers of pegs in the board.

    Returns:
        QuantumCircuit: The complete QGB circuit, ready for simulation.
    """
    if layers < 1:
        raise ValueError("Number of layers must be at least 1.")

    # --- 1. Define Circuit Parameters ---
    num_output_bins = layers + 1
    num_board_qubits = 2 * layers + 1

    q_control = QuantumRegister(1, 'control')
    q_board = QuantumRegister(num_board_qubits, 'board')
    c_output = ClassicalRegister(num_output_bins, 'c')
    qc = QuantumCircuit(q_control, q_board, c_output)

    # --- 2. Build the Quantum Galton Board Circuit ---

    # Start by placing the 'ball' on the middle qubit of the top row.
    initial_ball_qubit_index = layers
    qc.x(q_board[initial_ball_qubit_index])
    qc.barrier()

    # Loop through each layer of the board
    for l in range(1, layers + 1):
        print(f"Building Layer {l}...")
        # Reset the control qubit and apply H-gate ONCE for the new layer.
        qc.reset(q_control[0])
        qc.h(q_control[0])
        qc.barrier(label=f"Start L{l}")

        # Loop through each peg within the current layer
        num_pegs_in_layer = l
        for p in range(num_pegs_in_layer):
            # Calculate the indices for the interacting qubits for this peg
            ball_in_idx = layers - (l - 1) + (2 * p)
            left_out_idx = ball_in_idx - 1
            right_out_idx = ball_in_idx + 1

            # Apply the quantum peg logic
            quantum_peg_logic(qc, q_control[0], q_board[ball_in_idx], q_board[left_out_idx], q_board[right_out_idx])

            # Add the corrective CNOT gate between pegs in the same row
            is_last_peg_in_row = (p == num_pegs_in_layer - 1)
            if not is_last_peg_in_row:
                # The control is the qubit shared between the current peg and the next.
                cnot_control_idx = right_out_idx
                qc.cx(q_board[cnot_control_idx], q_control[0])
        qc.barrier(label=f"End L{l}")


    # --- 3. Measure the Final Output Bins ---
    # The final positions of the 'ball' are the qubits at even indices.
    for i in range(num_output_bins):
        measurement_qubit_idx = 2 * i
        qc.measure(q_board[measurement_qubit_idx], c_output[i])

    return qc

# --- Main execution block ---
if __name__ == '__main__':
    # Define the number of layers for our Galton Board
    NUM_LAYERS = 4

    # Create the circuit
    qgb_circuit = create_qgb_circuit(NUM_LAYERS)

    # --- 4. Display and Simulate the Circuit ---
    print(f"\n--- Quantum Circuit for {NUM_LAYERS}-Layer Galton Board ---")
    display(qgb_circuit.draw('mpl', style='clifford', fold=-1))

    print("\n--- Simulating the Circuit (using V2 Primitives) ---")
    sampler = StatevectorSampler()
    job = sampler.run([qgb_circuit], shots=8192)
    result = job.result()
    counts = result[0].data.c.get_counts()

    # --- 5. Verify and Plot the Results ---
    print("\n--- Simulation Results ---")
    # The raw results are binary strings. We remap them to bin numbers for clarity.
    num_output_bins = NUM_LAYERS + 1
    remapped_counts = {}
    for i in range(num_output_bins):
        # Create the expected binary string for this bin (e.g., '00100' for bin 2 of 5)
        binary_string = '0' * (num_output_bins - 1 - i) + '1' + '0' * i
        remapped_counts[f"Bin {i}"] = counts.get(binary_string, 0)

    print(remapped_counts)

    # The plot should show a binomial distribution, approximating a Gaussian curve.
    # For 4 layers, the expected ratio is 1:4:6:4:1.
    plot_histogram(remapped_counts, title=f'{NUM_LAYERS}-Layer QGB Simulation Results')
    plt.xlabel("Output Bin")
    plt.ylabel("Frequency (Counts)")
    plt.show()
```

```python
#
# General Quantum Galton Board Algorithm
#
# This script provides a function to generate a Quantum Galton Board
# circuit for any specified number of layers.
#
# This version contains the final corrected logic for the alternating
# gate chain in each layer, ensuring the correct number of CNOT gates.
#
# Qiskit Version: 1.4.2
#

import numpy as np
import matplotlib.pyplot as plt
from qiskit import QuantumCircuit, ClassicalRegister, QuantumRegister
from qiskit.primitives import StatevectorSampler
from qiskit.visualization import plot_histogram

# --- Main Function to Create the Galton Board Circuit ---
def create_qgb_circuit(layers):
    """
    Generates a Quantum Galton Board circuit for a given number of layers.

    Args:
        layers (int): The number of layers of pegs in the board.

    Returns:
        QuantumCircuit: The complete QGB circuit, ready for simulation.
    """
    if layers < 1:
        raise ValueError("Number of layers must be at least 1.")

    # --- 1. Define Circuit Parameters ---
    num_output_bins = layers + 1
    num_board_qubits = 2 * layers + 1

    q_control = QuantumRegister(1, 'control')
    q_board = QuantumRegister(num_board_qubits, 'board')
    c_output = ClassicalRegister(num_output_bins, 'c')
    qc = QuantumCircuit(q_control, q_board, c_output)

    # --- 2. Build the Quantum Galton Board Circuit ---

    # Start by placing the 'ball' on the middle qubit of the top row.
    initial_ball_qubit_index = layers
    qc.x(q_board[initial_ball_qubit_index])
    qc.barrier()

    # Loop through each layer of the board
    for l in range(1, layers + 1):
        print(f"Building Layer {l}...")
        # Reset the control qubit and apply H-gate ONCE for the new layer.
        qc.reset(q_control[0])
        qc.h(q_control[0])
        qc.barrier(label=f"Start L{l}")

        # *** FINAL CORRECTED CORE LOGIC ***
        
        # 1. Determine the start and end qubits for this layer's path.
        start_qubit = layers - l
        end_qubit = layers + l
        
        # 2. Build the alternating chain of CSWAP and CX gates.
        # For a layer with 2*l CSWAPs, we need exactly 2*l - 1 CXs.
        for i in range(start_qubit, end_qubit):
            # Apply CSWAP between the current qubit and the next one in the path.
            qc.cswap(q_control[0], q_board[i], q_board[i+1])
            
            # Apply the corrective CX gate, but NOT after the very last CSWAP.
            if i < end_qubit - 1:
                qc.cx(q_board[i+1], q_control[0])
        
        qc.barrier(label=f"End L{l}")


    # --- 3. Measure the Final Output Bins ---
    # The final positions of the 'ball' are the qubits at even indices.
    for i in range(num_output_bins):
        measurement_qubit_idx = 2 * i
        qc.measure(q_board[measurement_qubit_idx], c_output[i])

    return qc

# --- Main execution block ---
if __name__ == '__main__':
    # Define the number of layers for our Galton Board
    NUM_LAYERS = 4

    # Create the circuit
    qgb_circuit = create_qgb_circuit(NUM_LAYERS)

    # --- 4. Display and Simulate the Circuit ---
    print(f"\n--- Quantum Circuit for {NUM_LAYERS}-Layer Galton Board ---")
    display(qgb_circuit.draw('mpl', style='clifford', fold=-1))

    print("\n--- Simulating the Circuit (using V2 Primitives) ---")
    sampler = StatevectorSampler()
    job = sampler.run([qgb_circuit], shots=8192)
    result = job.result()
    counts = result[0].data.c.get_counts()

    # --- 5. Verify and Plot the Results ---
    print("\n--- Simulation Results ---")
    # The raw results are binary strings. We remap them to bin numbers for clarity.
    num_output_bins = NUM_LAYERS + 1
    remapped_counts = {}
    for i in range(num_output_bins):
        # Create the expected binary string for this bin (e.g., '00100' for bin 2 of 5)
        binary_string = '0' * (num_output_bins - 1 - i) + '1' + '0' * i
        remapped_counts[f"Bin {i}"] = counts.get(binary_string, 0)

    print(remapped_counts)

    # The plot should show a binomial distribution, approximating a Gaussian curve.
    # For 4 layers, the expected ratio is 1:4:6:4:1.
    plot_histogram(remapped_counts, title=f'{NUM_LAYERS}-Layer QGB Simulation Results')
    plt.xlabel("Output Bin")
    plt.ylabel("Frequency (Counts)")
    plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from qiskit import QuantumCircuit, ClassicalRegister, QuantumRegister
from qiskit.primitives import StatevectorSampler
from qiskit.visualization import plot_histogram
from math import comb

def create_qgb_circuit_correct(layers):
    """
    Quantum Galton Board implementation strictly following arXiv:2202.01735v1
    """
    if layers < 1:
        raise ValueError("Number of layers must be at least 1.")

    # Paper specifies: need one control qubit + (2*layers + 1) board qubits
    num_board_qubits = 2 * layers + 1
    num_output_bins = layers + 1

    q_control = QuantumRegister(1, 'ctrl')
    q_board = QuantumRegister(num_board_qubits, 'board')
    c_output = ClassicalRegister(num_output_bins, 'c')
    qc = QuantumCircuit(q_control, q_board, c_output)

    # Step 1: Initialize the ball at the center position
    center_pos = layers
    qc.x(q_board[center_pos])
    qc.barrier(label="Initialize")

    # Step 2: Apply the quantum walk operations layer by layer
    for layer in range(1, layers + 1):
        print(f"Processing layer {layer}...")
        
        # Reset control qubit for new layer (except first)
        if layer > 1:
            qc.reset(q_control[0])
        
        # Apply Hadamard to create superposition
        qc.h(q_control[0])
        qc.barrier(label=f"Layer {layer}")
        
        # Apply the correct peg operations for this layer
        # For layer l, we need to process 2*l positions
        start_pos = layers - layer
        end_pos = layers + layer
        
        # Apply CSWAP operations followed by corrective CNOTs
        for pos in range(start_pos, end_pos + 1):
            if pos < end_pos:  # Not the last position
                # CSWAP between adjacent positions
                qc.cswap(q_control[0], q_board[pos], q_board[pos + 1])
                
                # Corrective CNOT (except after the last CSWAP in the layer)
                if pos < end_pos - 1:
                    qc.cx(q_board[pos + 1], q_control[0])

    # Step 3: Measure all board qubits to determine final ball position
    qc.barrier(label="Measure")
    for i in range(num_output_bins):
        # Measure positions that correspond to output bins
        measure_pos = 2 * i
        if measure_pos < num_board_qubits:
            qc.measure(q_board[measure_pos], c_output[i])

    return qc

def create_qgb_circuit_paper_exact(layers):
    """
    Alternative implementation following the paper's Figure 4 more precisely
    """
    if layers < 1:
        raise ValueError("Number of layers must be at least 1.")

    num_board_qubits = 2 * layers + 1
    num_output_bins = layers + 1

    q_control = QuantumRegister(1, 'ctrl')
    q_board = QuantumRegister(num_board_qubits, 'board')
    c_output = ClassicalRegister(num_output_bins, 'c')
    qc = QuantumCircuit(q_control, q_board, c_output)

    # Initialize ball at center
    center = layers
    qc.x(q_board[center])
    qc.barrier(label="Init")

    # Process each layer according to the paper's exact specification
    for layer in range(1, layers + 1):
        print(f"Building layer {layer}...")
        
        # Reset and prepare control
        if layer > 1:
            qc.reset(q_control[0])
        qc.h(q_control[0])
        qc.barrier(label=f"L{layer}_start")
        
        # For layer i, apply quantum walk over 2*i+1 positions
        layer_width = 2 * layer + 1
        start_idx = layers - layer
        
        # Apply the CSWAP chain with proper CNOTs
        for j in range(layer_width - 1):
            pos = start_idx + j
            # CSWAP between consecutive positions
            qc.cswap(q_control[0], q_board[pos], q_board[pos + 1])
            
            # Add CNOT for interference (not after last CSWAP)
            if j < layer_width - 2:
                qc.cx(q_board[pos + 1], q_control[0])
        
        qc.barrier(label=f"L{layer}_end")

    # Measure output positions
    qc.barrier(label="Measure")
    for i in range(num_output_bins):
        output_pos = 2 * i
        qc.measure(q_board[output_pos], c_output[i])

    return qc

def analyze_qgb_results(counts, num_layers):
    """
    Analyze QGB results and extract bin distributions
    """
    num_bins = num_layers + 1
    bin_counts = {f"Bin {i}": 0 for i in range(num_bins)}
    
    for bitstring, count in counts.items():
        # Find which bit is set (one-hot encoding)
        for i, bit in enumerate(reversed(bitstring)):
            if bit == '1':
                if i < num_bins:
                    bin_counts[f"Bin {i}"] += count
                break
    
    return bin_counts

def expected_binomial_distribution(n):
    """Calculate expected binomial coefficients for n layers"""
    return [comb(n, k) for k in range(n + 1)]

# Test both implementations
if __name__ == '__main__':
    NUM_LAYERS = 4
    
    print(f"Testing CORRECTED {NUM_LAYERS}-layer Quantum Galton Board")
    print("Implementation following arXiv:2202.01735v1 precisely")
    
    # Try the paper-exact implementation
    qgb_circuit = create_qgb_circuit_paper_exact(NUM_LAYERS)
    
    print(f"\n--- Circuit Statistics ---")
    print(f"  Total Qubits: {qgb_circuit.num_qubits}")
    print(f"  Board Qubits: {2 * NUM_LAYERS + 1}")
    print(f"  Circuit Depth: {qgb_circuit.depth()}")
    print(f"  Gate Count: {qgb_circuit.count_ops()}")
    
    # Display circuit
    print(f"\n--- Circuit Diagram ---")
    display(qgb_circuit.draw('mpl', style='iqp', fold=30))
    
    # Simulate
    print(f"\n--- Simulation Results ---")
    sampler = StatevectorSampler()
    job = sampler.run([qgb_circuit], shots=8192)
    result = job.result()
    counts = result[0].data.c.get_counts()
    
    print(f"Raw measurement results:")
    for bitstring, count in sorted(counts.items()):
        print(f"  {bitstring}: {count}")
    
    # Analyze results
    bin_counts = analyze_qgb_results(counts, NUM_LAYERS)
    total_shots = sum(bin_counts.values())
    
    print(f"\nBin distribution:")
    for bin_name, count in bin_counts.items():
        percentage = (count / total_shots) * 100 if total_shots > 0 else 0
        print(f"  {bin_name}: {count} ({percentage:.1f}%)")
    
    # Expected distribution
    expected_coeffs = expected_binomial_distribution(NUM_LAYERS)
    total_expected = sum(expected_coeffs)
    
    print(f"\nExpected binomial distribution:")
    print(f"  Coefficients: {expected_coeffs}")
    print(f"  Normalized percentages:")
    for i, coeff in enumerate(expected_coeffs):
        expected_percentage = (coeff / total_expected) * 100
        expected_count = (coeff / total_expected) * total_shots
        print(f"    Bin {i}: {coeff} ({expected_percentage:.1f}%, ~{expected_count:.0f} counts)")
    
    # Enhanced visualization
    fig, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(18, 6))
    
    # Actual results
    bins = list(range(NUM_LAYERS + 1))
    actual_counts = [bin_counts[f"Bin {i}"] for i in bins]
    
    ax1.bar(bins, actual_counts, alpha=0.8, color='skyblue', edgecolor='navy')
    ax1.set_title(f'Actual Results ({NUM_LAYERS} layers)', fontsize=12, fontweight='bold')
    ax1.set_xlabel('Output Bin')
    ax1.set_ylabel('Count')
    ax1.grid(True, alpha=0.3)
    
    # Expected vs Actual comparison
    expected_normalized = [(c/total_expected)*total_shots for c in expected_coeffs]
    x = np.arange(len(bins))
    width = 0.35
    
    ax2.bar(x - width/2, actual_counts, width, label='Actual', alpha=0.8, color='skyblue')
    ax2.bar(x + width/2, expected_normalized, width, label='Expected', alpha=0.8, color='lightcoral')
    ax2.set_title('Actual vs Expected Distribution', fontsize=12, fontweight='bold')
    ax2.set_xlabel('Output Bin')
    ax2.set_ylabel('Count')
    ax2.legend()
    ax2.set_xticks(x)
    ax2.set_xticklabels([f'Bin {i}' for i in bins])
    ax2.grid(True, alpha=0.3)
    
    # Percentage error
    percentage_errors = []
    for i in range(len(actual_counts)):
        if expected_normalized[i] > 0:
            error = abs(actual_counts[i] - expected_normalized[i]) / expected_normalized[i] * 100
            percentage_errors.append(error)
        else:
            percentage_errors.append(0)
    
    ax3.bar(bins, percentage_errors, alpha=0.8, color='orange', edgecolor='red')
    ax3.set_title('Percentage Error by Bin', fontsize=12, fontweight='bold')
    ax3.set_xlabel('Output Bin')
    ax3.set_ylabel('Percentage Error (%)')
    ax3.grid(True, alpha=0.3)
    
    plt.tight_layout()
    plt.show()
    
    # Statistical analysis
    try:
        from scipy.stats import chisquare
        if total_shots > 0 and all(expected_normalized):
            chi2_stat, p_value = chisquare(actual_counts, expected_normalized)
            print(f"\nGoodness of fit test:")
            print(f"  Chi-square statistic: {chi2_stat:.4f}")
            print(f"  P-value: {p_value:.6f}")
            print(f"  Result: {'PASS' if p_value > 0.05 else 'FAIL'} (=0.05)")
            
            # Additional analysis
            mean_error = np.mean(percentage_errors)
            max_error = np.max(percentage_errors)
            print(f"\nError Analysis:")
            print(f"  Mean percentage error: {mean_error:.2f}%")
            print(f"  Maximum percentage error: {max_error:.2f}%")
            
    except ImportError:
        print("\nScipy not available for statistical test")
    except Exception as e:
        print(f"\nStatistical test error: {e}")

    print(f"\nExpected ratio for {NUM_LAYERS} layers: {':'.join(map(str, expected_coeffs))}")
    
    # Debug: Check if we're getting the right number of measurements
    total_measured = sum(counts.values())
    print(f"\nDebug Info:")
    print(f"  Total measurements recorded: {total_measured}")
    print(f"  Expected total shots: 8192")
    print(f"  Measurement efficiency: {(total_measured/8192)*100:.1f}%")
```
