```python
%matplotlib inline
from qiskit import QuantumCircuit, transpile
from qiskit.circuit.library import CSwapGate
from qiskit_aer import AerSimulator
from qiskit.visualization import plot_histogram
import matplotlib.pyplot as plt

# 1 coin qubit + 3 position qubits, 3 classical bits
qc = QuantumCircuit(4, 3)  # q[0]=coin, q[1:4]=positions

# Initialize ball at center position (q2)
qc.x(2)

# Put coin qubit in superposition
qc.h(0)

# First CSWAP (coin as control, swap q1 and q2)
qc.append(CSwapGate(), [0, 1, 2])

# Inverted CNOT (control q2, target coin, active when q2==|0⟩)
qc.cx(2, 0, ctrl_state=0)

# Second CSWAP — inverted control so swap happens when coin==|0⟩
qc.x(0)  # invert coin control
qc.append(CSwapGate(), [0, 2, 3])
qc.x(0)  # restore coin

# Measure position qubits
qc.measure(1, 0)
qc.measure(2, 1)
qc.measure(3, 2)

# Show circuit
qc.draw('mpl')
plt.show()

# Simulate
sim = AerSimulator()
compiled = transpile(qc, sim)
result = sim.run(compiled, shots=1000).result()
counts = result.get_counts()

# Plot results
plot_histogram(counts)
plt.title("1-layer QGB (only 001 or 100 expected)")
plt.show()

# Print counts
print("Measurement results:")
for state, count in counts.items():
    print(f"State {state}: {count} times")
```

```python
from qiskit import QuantumCircuit, transpile
from qiskit.circuit.library import CSwapGate
from qiskit_aer import AerSimulator
from qiskit.visualization import plot_histogram
import matplotlib.pyplot as plt

# 1 coin qubit + 3 position qubits, 3 classical bits
qc = QuantumCircuit(4, 3)  # q[0]=coin, q[1:4]=positions

# === Initial state ===
qc.x(2)  # Ball starts in middle position q2

# === Step 1: put coin in superposition ===
qc.h(0)

# === Step 2: coin decides left swap (middle ↔ left) ===
qc.append(CSwapGate(), [0, 1, 2])

# === Step 3: inverted CNOT — triggers if middle is empty ===
qc.x(2)
qc.cx(2, 0)  # now control is effectively q2 == 0
qc.x(2)

# === Step 4: coin decides right swap (middle ↔ right) ===
qc.append(CSwapGate(), [0, 2, 3])

# === Measure position qubits ===
qc.measure(1, 0)  # left
qc.measure(2, 1)  # middle
qc.measure(3, 2)  # right

# Show circuit diagram
qc.draw('mpl')
plt.show()

# === Simulation ===
sim = AerSimulator()
compiled = transpile(qc, sim)
result = sim.run(compiled, shots=1000).result()
counts = result.get_counts()

# Show result
plot_histogram(counts)
plt.title("1-layer QGB")
plt.show()

# Print the counts to verify
print("Measurement results:")
for state, count in counts.items():
    print(f"State {state}: {count} times")
```
